```python
import re
import nltk
from nltk.corpus import stopwords
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

# Mengunduh daftar stopwords
nltk.download('stopwords')

# Mendapatkan daftar stopwords bahasa Indonesia
stop_words = set(stopwords.words('indonesian'))

# Membuat stemmer bahasa Indonesia dengan Sastrawi
factory = StemmerFactory()
stemmer = factory.create_stemmer()

def preprocess_text(text):
    # Hapus karakter khusus, angka, dan URL
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)
    text = re.sub(r'\@\w+|\#', '', text)
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'[^\w\s]', '', text)
    text = text.lower()  # Case folding

    # Hapus stop words
    text = ' '.join([word for word in text.split() if word not in stop_words])

    # Stemming dengan Sastrawi
    text = ' '.join([stemmer.stem(word) for word in text.split()])
    return text

# Aplikasi fungsi preprocess_text pada kolom 'text' DataFrame
df['cleaned_content'] = df['content'].apply(preprocess_text)
df.head()
```

```python
from googletrans import Translator

translator = Translator()

def translate_text(text):
    try:
        translation = translator.translate(text, dest='en')  # Translate to English
        return translation.text.lower()  # Convert translation to lowercase
    except Exception as e:
        return text.lower()  # Return original text in lowercase if translation fails

# Apply the translation function to the 'cleaned_content' column
df['translated_content'] = df['cleaned_content'].apply(translate_text)
df.head()
```

```python
import pandas as pd
from textblob import TextBlob

# Asumsikan df sudah ada dan memiliki kolom 'translated_content'

def analyze_sentiment(text):
    # Menggunakan TextBlob untuk menghitung polaritas teks
    blob = TextBlob(text)
    polarity = blob.sentiment.polarity

    # Mengklasifikasikan sentimen berdasarkan polaritas
    if polarity > 0:
        return 'Positif'
    elif polarity == 0:
        return 'Netral'
    else:
        return 'Negatif'

# Aplikasi analisis sentimen pada kolom 'translated_content'
df['sentiment'] = df['translated_content'].apply(analyze_sentiment)
df.head()
```
